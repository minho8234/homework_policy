```python
# Load
from langchain_community.document_loaders import PyPDFLoader
from langchain_community.vectorstores import FAISS
from langchain_community.embeddings import HuggingFaceEmbeddings

loader = PyPDFLoader('./2026.pdf')
docs = loader.load()
print(len(docs))
for doc in docs:
    print(doc.metadata)
    print(doc.page_content)
    print('\n\n')
```

```python
# RecursiveCharacterTextSplitter
from langchain_text_splitters import RecursiveCharacterTextSplitter

text = """
당신은 30년 경력의 대한민국 재정경제부 정책담당관입니다. 사용자의 질문에는 반드시 제공된 Context(문맥) 안에서만 답변해주십시오. Context에 명확히 확인되지 않는 내용에 대해선 "모른다"고 따뜻하게 안내해주시고, 추측이나 상상은 절대 하지 마십시오. 항상 사용자 반응에 호응하며 친절하고 따뜻한 말투로 안내하십시오. 답변, 설명, 예시는 모두 한글로 작성해주십시오.

- 답변 시 반드시 주어진 Context를 바탕으로 논리적 사고와 설명(Reasoning)을 먼저 제시하고, 그 후 결론이나 요약(Conclusions)을 마지막에 제시해야 합니다.
- Context에 없는 내용이라면 내용을 알수없다고 하지말고 검색을 통해 내용을 꼭 답변을 해줘야합니다.
- 사용자가 추가로 궁금해 하면 언제든 문의하라는 등 적극적이고 따뜻하며 신뢰감을 주는 어투를 유지하십시오.
- 질문에 답변을 못할 때도 따뜻한 어조로 안내를 해주세요.
- 내용 전달은 짧고 명확하게 (2~4문장), reasoning(설명 과정)이 반드시 먼저, 결론 혹은 요약이 반드시 마지막에 오도록 작성하십시오.
- 모든 답변은 한글로만 작성하세요.
- Context를 우선적으로 꼼꼼히 확인하고, 부정확하거나 정보가 없을 경우 따로 언급하지 마세요.

# 답변 작성 순서

1. Context를 바탕으로 사용자의 질문에 어떻게 접근할지 논리적으로 설명(Reasoning)합니다.
    Context에 없을경우 검색 2번이상 확인후 정확한 내용을 설명해줘야합니다.
2. 설명 이후 답변의 결론이나 요약(Conclusion, Summary)을 마지막에 제시합니다.

# 출력 형식

- 사용자의 질문에 대해 직접 대화하듯 답변하세요.
- 항상 reasoning(설명)의 과정을 먼저, 결론 및 요약을 마지막에 두세요.
- 각 답변은 2~4문장 내로 reasoning과 결론을 모두 담아주세요.
- 반드시 한글로만 작성하십시오.

# 예시

**사용자 질문:** "올해 소상공인 지원 정책이 어떻게 변경될 예정인가요?"

**주어진 Context:** "2024년 재정경제부는 소상공인 대상 대출 지원을 확대할 계획입니다."

**답변 예시:**
"주어진 Context를 살펴보면 2024년에 재정경제부가 소상공인 대상 대출 지원을 확대할 계획이라고 밝혔습니다. 따라서 올해 소상공인분들께서는 대출 지원을 더 많이 받으실 수 있을 것으로 예상됩니다. 또 궁금한 점 있으시면 언제든 말씀해주세요!"

**사용자 질문:** "세제 개편 내용도 알려주세요!"

**주어진 Context:** "2024년 재정경제부는 소상공인 대상 대출 지원을 확대할 계획입니다."

**답변 예시:**
"Context 정보를 확인해보았으나, 세제 개편에 관한 내용은 포함되어 있지 않습니다. 해당 부분은 Context에서 알 수 없어 안내가 어려운 점 양해 부탁드립니다."

(실제 답변은 사용자의 질문과 Context에 따라 예시보다 더 길거나, 짧거나, 다양한 세부 정보가 포함될 수 있습니다.)

# 중요 안내

- reasoning이 반드시 먼저, 결론이 반드시 나중에 오도록 답변 구조를 유지해야 합니다.
- Context에서 확인되지 않는 정보는 절대 유추하거나 추가하지 말고, 검색 2번이상 확인후 정확한 내용을 설명해줘야합니다.
- 항상 한글로 설명 및 안내하세요.

(위 지침과 예시, 답변 순서 모두 반드시 철저히 준수하시기 바랍니다.)
'''),
    ('human', '''
사용자의 질문을 파악하고, 다음 Context의 내용을 꼼꼼히 살펴본후, 대답해주세요.

Question:
{query}

Context:
{context}
''')
])
"""


splitter = RecursiveCharacterTextSplitter(
    chunk_size=100,
    chunk_overlap=30,
    separators=['\n\n', '\n', '.', ' ', '']
)
chunks = splitter.split_text(text)
for i, chunk in enumerate(chunks):
    print(f'{i} ({len(chunk)}): {chunk}')
```

```python
# Split
splitter = RecursiveCharacterTextSplitter(
    chunk_size=200,
    chunk_overlap=60,
    separators=['\n\n','\n', ' ', '']
)
documents = splitter.split_documents(docs)
print(len(documents))

for i, doc in enumerate(documents):
    print(f'{i}:(글자길이: {len(doc.page_content)}) (PDF Page: {doc.metadata['page_label']})')
    print(doc.page_content)
    print()
```

```python
# Embed
# - Embedding모델 openai
# - Vector DB (chroma)
from langchain_openai import OpenAIEmbeddings
from langchain_chroma import Chroma

embedding_model = OpenAIEmbeddings(model='text-embedding-3-small') # 1536
vector_store = Chroma.from_documents(documents=documents, embedding=embedding_model)
vector_store
```

```python
# 벡터서치 테스트
# - vector_store.similarity_search()
# - vector_store_retriver.invoke()
query = '장애인고용개선장려금은 지급받으려면 조건이 어떻게 돼?' # 1536차원 벡터 변환
# retrievals = vector_store.similarity_search(query, k=4)
retrievals = vector_store.similarity_search_with_score(query, k=4) # L2 유클리디안거리 측정(0에 가까울수록 좋음)

for doc, score in retrievals:
    print(f'Score: {score:.4f} / {doc.metadata['page_label']} Page')
    print(doc.page_content)
    print()
```

```python
vector_store_retriever = vector_store.as_retriever(
    search_type='similarity',
    search_kwargs={'k': 4}
)

retrievals = vector_store_retriever.invoke(query)

for doc in retrievals:
    print(f'{doc.metadata['page_label']} Page:')
    print(doc.page_content)
    print()
```

```python
# 프롬프트 준비
from langchain_core.prompts import ChatPromptTemplate

prompt = ChatPromptTemplate.from_messages([
    ('system', '''
당신은 대한민국 기획재정부(구 재정경제부)에서 30년간 근무하며 국민을 위해 헌신해온 베테랑 '정책담당관'입니다.
당신의 임무는 제공된 [Context]를 바탕으로 국민(사용자)의 질문에 대해 가장 정확하고 따뜻한 답변을 제공하는 것입니다.

# 핵심 원칙
1. **Context 우선 및 검색 보완**: 우선적으로 제공된 [Context]를 분석하십시오. 만약 Context에 정보가 부족하거나 없다면, 즉시 웹 검색을 통해 최신 정책 정보를 확인하여 답변에 포함하십시오. "모른다"는 말 대신 적극적으로 정보를 찾아 알려주십시오.
2. **논리적 구성 (Reasoning First)**: 답변을 할 때는 반드시 [검토 및 분석] 과정을 먼저 서술하고, 그 뒤에 [상세 답변]을 제시하십시오.
3. **가독성 극대화**: 줄글로 길게 늘어놓지 말고, 소제목, 불릿 포인트(•), 번호 매기기 등을 활용하여 내용을 한눈에 파악할 수 있도록 작성하십시오.
4. **따뜻한 전문가의 태도**: 30년 경력의 연륜이 묻어나는 신뢰감 있고 친절한 어조를 유지하십시오.
5. **문의처 필수 포함**: 정책과 관련된 담당 부서명과 전화번호를 확인하여 답변 하단에 반드시 기재해주십시오.

# 답변 작성 구조 (반드시 준수)

**1. [검토 및 분석]**
- 사용자의 질문 의도를 파악하고, Context 및 검색된 정보와의 연관성을 2~3문장으로 논리적으로 설명하십시오.

**2. [상세 정책 안내]**
- 정책의 핵심 내용, 변경 사항, 지원 대상 등을 이해하기 쉽게 항목별로 나누어 정리하십시오.
- (예시 형식)
  • **지원 내용:** ...
  • **신청 대상:** ...
  • **신청 기간:** ...

**3. [요약 및 제언]**
- 전체 내용을 한 문장으로 요약하고, 전문가로서의 따뜻한 조언을 덧붙이십시오.

**4. [자세한 문의]**
- 관련 부서: OOO과
- 전화번호: 044-000-0000 (또는 관련 콜센터)
- 홈페이지: (확인될 경우 기재)

위 형식을 엄격히 준수하여 한국어로 답변하십시오.
'''),
    ('human', '''
아래 제공된 [Context]와 사용자의 [Question]을 면밀히 검토하여 답변해 주십시오.

[Context]:
{context}

[Question]:
{query}
''')
])

query = '장애인고용개선장려금은 지급받으려면 조건이 어떻게 돼?'
retrievals = vector_store_retriever.invoke(query)
context = '\n\n'.join(doc.page_content for doc in retrievals)

prompt_value = prompt.invoke({'query': query, 'context': context})
print(prompt_value.messages)

print(prompt_value.messages[0].content)
print(prompt_value.messages[1].content)
```

```python
from google.colab import userdata
import os

os.environ['PPLX_API_KEY'] = userdata.get('PPLX_API_KEY')
```

```python
# LLM Chain (참조문서 내용 포함)
from langchain.chat_models import init_chat_model
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnablePassthrough

model = init_chat_model('perplexity:sonar')
output_parser = StrOutputParser()

def format_docs_with_metadata(docs):
    formatted = []
    for doc in docs:
        source = doc.metadata.get('source', 'Unknown')
        page_label = doc.metadata.get('page_label', 'Unknown')
        page_content = doc.page_content
        text = f'''
[Source/Page Label]
{source}/{page_label}
[Content]
{page_content}
'''
        formatted.append(text)

    return '\n\n'.join(formatted)

chain = (
    {'query': RunnablePassthrough(), 'context': vector_store_retriever | format_docs_with_metadata}
    | prompt
    | model
    | output_parser
)


query = '26년 신생아 지급 정책은 어떤것들이 있어?'
print(chain.invoke(query))
```
